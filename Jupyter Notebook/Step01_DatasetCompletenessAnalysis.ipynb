#### Global Sampling Rate
In our study, the global sampling rate is computed by first summing the total number of samples and the total duration across all sessions for each device and signal. The sampling rate is then calculated by dividing the total samples by the total duration. Finally, this rate is compared to the expected sampling rate to determine the data coverage.

```python
result_dir = Path('../../SamplingRateAnalysis')
sampling_summary_df = pd.read_csv(f"{result_dir}/global_sampling_statistics.csv")
display(sampling_summary_df)
sampling_summary_df_std = pd.read_csv(f"{result_dir}/sampling_rate_summary.csv")
display(sampling_summary_df_std)
```
