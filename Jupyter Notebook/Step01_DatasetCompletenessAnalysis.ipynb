```python
from Step01_DatasetCompletenessAnalysis.SamplingRate import OverallSamplingRateAnalyzer
```

```python
# Initialize the Overall Sampling Rate Analyzer
analyzer = OverallSamplingRateAnalyzer()
```

```python
# Specify the participant ID
participant_id = "P03"

# Process the data for the specified participant
result = analyzer.process_participant(participant_id)
```

```python
# Display any missing files or column errors
print(f"Participant: {participant_id}")
print("Missing Files:", result.get('missing_files', []))
print("Column Errors:", result.get('column_errors', []))

# Display session-wise sampling rate results
for session, session_data in result['sessions'].items():
    print(f"\nSession: {session}")
    for device, device_data in session_data.items():
        print(f"  Device: {device}")
        print(f"    Duration (s): {device_data['duration']}")
        for signal, stats in device_data['signals'].items():
            if 'error' in stats:
                print(f"    {signal.upper()}: {stats['error']}")
            else:
                print(f"    {signal.upper()}:")
                print(f"      Expected Rate: {stats['expected_rate']} Hz")
                print(f"      Actual Rate: {stats['actual_rate']:.2f} Hz")
                print(f"      Expected Samples: {stats['expected_samples']}")
                print(f"      Actual Samples: {stats['actual_samples']}")
                print(f"      Coverage Ratio: {stats['coverage_ratio']:.2%}")
```
